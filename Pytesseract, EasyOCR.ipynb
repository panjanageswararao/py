# Extract Text From PDF

```python
import io
```

```python
from PIL import Image
import pytesseract
from wand.image import Image as wi
```

```python
pytesseract.pytesseract.tesseract_cmd = r'c:\Users\Panja\AppData\Local\Programs\Tesseract-OCR\\tesseract.exe'
```

```python
pdf = wi(filename = "demo-invoice-1.pdf",resolution=300)
pdfImg = pdf.convert('jpeg')
```

```python
imgBlobs = []
```

```python
for img in pdfImg.sequence:
    page = wi(image = img)
    imgBlobs.append(page.make_blob('jpeg'))
```

```python
extracted_text =[]
```

```python
for imgBlog in imgBlobs:
    im = Image.open(io.BytesIO(imgBlob))
    text = pytesseract.image_to_string(im, lang='eng')
    extracted_text.append(text)
```

```python

print(extracted_text[0])
```

# Extract Text From Images Using EasyOCR

```python
import easyocr
```

```python
import matplotlib.pyplot as plt
import cv2
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8,16
```

```python
reader = easyocr.Reader(['en'])
```

```python
Image("Image1.jpg")
```

```python
reader.readtext('Image1.jpg', detail = 0)
```

```python
output = reader.readtext('Image1.jpg')
```

```python
output
```

```python
cord = output[-1][0]
```

```python
x_min, y_min = [int(min(idx)) for idx in zip(*cord)]

x_max, y_max = [int(max(idx)) for idx in zip(*cord)]
```

```python
image = cv2.imread('Image1.jpg')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
